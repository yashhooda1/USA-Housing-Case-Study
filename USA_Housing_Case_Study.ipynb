## Data cleaning

### Subtask:
Clean the dataset by handling outliers and duplicates.


**Reasoning**:
Clean the dataset by handling outliers and duplicates based on the exploratory data analysis.  I will use the IQR method to detect and cap outliers in the numerical features (excluding 'Price').  I will then remove duplicate rows.



```python
import pandas as pd
import numpy as np

# Identify outliers using IQR and cap them
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.clip(df[column], lower_bound, upper_bound)
    return df

numerical_cols = ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population']
for col in numerical_cols:
    df = cap_outliers(df, col)

# Remove duplicates
df_cleaned = df.drop_duplicates()

# Display the first few rows of the cleaned DataFrame
display(df_cleaned.head())
print(df_cleaned.shape)
```

## Model evaluation

### Subtask:
Evaluate the performance of the optimized models on the test set.


**Reasoning**:
Load the saved models and evaluate their performance on the test set.



```python
import pickle
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd

# Load the saved models
try:
    filename_linear = 'finalized_linear_model.sav'
    loaded_linear_model = pickle.load(open(filename_linear, 'rb'))

    filename_rf = 'finalized_rf_model.sav'
    loaded_rf_model = pickle.load(open(filename_rf, 'rb'))

    filename_gb = 'finalized_gb_model.sav'
    loaded_gb_model = pickle.load(open(filename_gb, 'rb'))
except FileNotFoundError:
    print("Error: One or more saved model files not found.")
    loaded_linear_model = None
    loaded_rf_model = None
    loaded_gb_model = None


# Evaluate the models
models = {
    'Linear Regression': loaded_linear_model,
    'Random Forest': loaded_rf_model,
    'Gradient Boosting': loaded_gb_model
}

results = {}

for model_name, model in models.items():
  if model:
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    results[model_name] = {
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2,
        'MAE': mae
    }
    print(f'{model_name} Evaluation Metrics:')
    print(f'MSE: {mse:.2f}')
    print(f'RMSE: {rmse:.2f}')
    print(f'R2: {r2:.2f}')
    print(f'MAE: {mae:.2f}')
  else:
    print(f"Skipping evaluation for {model_name} due to missing model file")


# Create a summary table
results_df = pd.DataFrame(results).T
display(results_df)

# Analyze and identify the best model (based on RMSE, R2)
print("\nAnalysis:")
if results:
    best_model = results_df['RMSE'].idxmin()
    print(f"The best performing model based on RMSE is: {best_model}")

    best_r2_model = results_df['R2'].idxmax()
    print(f"The best performing model based on R2 is: {best_r2_model}")
```
